Have to figure out the coordinate clusters because I can't download data automatically.  I have to do it manually and using zip codes.

```python
# I need a list of lists, top list is all the coordinates, sublist is ['lat','lon']
# first read the coordinate file
routeCoordDf = pd.read_csv('../input/runclustering/RouteUnitCoordinates.csv')
latlist = routeCoordDf['lat']
lonlist = routeCoordDf['lon']
coordlist = []
for x in range(len(latlist)) :
    coordlist.append([latlist.iloc[x],lonlist.iloc[x]])
print("doneA")
```

```python
from sklearn.cluster import KMeans
ckm = KMeans(n_clusters=20)
ckm.fit(coordlist)
ckm.cluster_centers_
```
